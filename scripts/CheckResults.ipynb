```python
import pandas as pd
import os
import numpy as np
```

```python
gt_path = "/data2/knn_datasets/sparse_datasets/msmarco_v1_passage/cocondenser/data/groundtruth.tsv"
```

```python
columns = ["query_id", "doc_id", "rank", "score"]
gt = pd.read_csv(gt_path, sep="\t", names=columns)
```

```python
grouped_gt = gt.groupby("query_id")
```

```python
results_dir = "../output_dir_partitioned/"
```

```python
for file in os.listdir(results_dir):
    path = os.path.join(results_dir, file)
    current_df = pd.read_csv(path, sep="\t", names=columns)
    intersection = 0
    n_docs = 0
    for (q_id, current_group) in current_df.groupby("query_id"):
        intersection += len(np.intersect1d(current_group.doc_id, grouped_gt.get_group(q_id).doc_id))
    print(f"File: {file}, Recall: {intersection / len(current_df)}")
```

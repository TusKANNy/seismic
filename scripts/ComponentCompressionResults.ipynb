```python
import pandas as pd
```

```python
import os
```

```python
base_dir = "../my_codec_logs/"
all = []
for item in filter(lambda x: x.endswith(".tsv"), os.listdir(base_dir)):
    all.append(pd.read_csv(os.path.join(base_dir, item), sep="\t"))
```

```python
all_df = pd.concat(all)
```

```python
all_df['avg_time_per_query_ms'] = all_df['avg_time_per_query_ms'] / 1000
```

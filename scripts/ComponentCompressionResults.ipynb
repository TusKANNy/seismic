```python
import pandas as pd
```

```python
import os
```

```python
def extract_results(base_dir):
    all = []
    for item in filter(lambda x: x.endswith(".tsv"), os.listdir(base_dir)):
        all.append(pd.read_csv(os.path.join(base_dir, item), sep="\t"))
    all_df = pd.concat(all)
    return all_df 
```

```python
assert len(pd.unique(all_df['compressed_memory_bytes'] - all_df['component_memory_bytes'])) == 1
values_mem_bytes = pd.unique(all_df['compressed_memory_bytes'] - all_df['component_memory_bytes'])[0]
```
